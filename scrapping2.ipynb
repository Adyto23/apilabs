```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://en.wikipedia.org/wiki/Python"
```

```python
response = requests.get(url)
```

```python
response.status_code # 200 status code means OK!
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select("li:nth-child(1) > a ")

#mw-content-text > div.mw-parser-output > ul:nth-child(2) > li:nth-child(1) > a
```

```python
soup.select("a")
```

```python
links=[]
num_iter = len(soup.select("a"))
tClist = soup.select("a")


for i in range(num_iter):
    links.append(tClist[i].get_text())
        
links
#mw-content-text > div.mw-parser-output > ul:nth-child(13) > li:nth-child(1) > a
```

```python
anchors = soup.find_all('a')
all_links = set()
for link in anchors:
    if(link.get('href') != '#'): 
        if 'Python' in link.get('href'):
            linkText = url+str(link.get('href'))
            all_links.add(link)
            print(linkText)
            
##first option 
```

```python
#[i for i in iterations]

thislist = []

for i in links:
    start_at=  str(i) 
    if 'Python' in i:
        link = "https://en.wikipedia.org/wiki/" + start_at
        link = link.replace(' ', '_')
        thislist.append(link)
        
thislist

data = pd.DataFrame({
    'links': thislist})

## second way , not giving all but most are right 
```
